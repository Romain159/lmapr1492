```python
from pymatgen.ext.matproj import MPRester
with MPRester("hAyTf8zq9AWy4PrFn") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-569779")

    # Dos for material id
    dos = m.get_dos_by_material_id("mp-569779")

    # Bandstructure for material id
    bandstructure = m.get_bandstructure_by_material_id("mp-569779")
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt

# Longueur d'onde du rayonnement X incident en Å
wavelength = 1.54060

# Création d'un objet XRDCalculator avec la longueur d'onde spécifiée
xrd_calc = XRDCalculator(wavelength=wavelength)

# Calcul du diffractogramme pour le matériau donné
pattern = xrd_calc.get_pattern(structure, scaled=True, two_theta_range=(0, 90))

# Affichage des trois premiers pics avec les indices hkl associés
for i in range(3):
    print("Pic", i+1, ":", pattern.hkls[i], "2θ =", pattern.x[i], "Intensité =", pattern.y[i])

# Tracé du diffractogramme
plt.plot(pattern.x, pattern.y, color='blue')
plt.xlabel('Angle de diffraction 2θ (degrés)')
plt.ylabel('Intensité (normalisée)')
plt.title('Diffractogramme pour le matériau mp-569779')
plt.show()
```
